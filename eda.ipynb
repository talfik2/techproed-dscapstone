### Object Columns

```python
# returning object cols
print("Object Columns: ", df.select_dtypes(include = ["object"]).columns)
```

* Since there is no null values in date and since we converted it to datetime format, we need to exclude it from `obj_df` that we are about to create.

```python
# Splitting object columns for encoding and naming as obj_df
obj_df = df[['iso_code', 'continent', 'location', 'tests_units']]
```

```python
from sklearn.impute import SimpleImputer
imr = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

imr = imr.fit(obj_df)
imputed_data = imr.transform(obj_df)
obj_df = pd.DataFrame(imputed_data)
obj_df.head(3)
```

```python
# Renaming Column Names
obj_df = obj_df.rename(columns={0:'iso_code', 1: 'continent', 2: 'location', 3: 'test_units'})
```

```python
obj_df.head(3)
```

```python
obj_df.isnull().values.any()
```

**Data Encoding**
* The computer reads numbers, even the most complex ones. It is much easier to understand numerical data rather than categorical, which is why I'll encode data.

```python
from sklearn.feature_extraction import DictVectorizer
# df'mizi dict'e dönüştürelim
df_dict = obj_df.to_dict("Records")
# Dict Vectorizer'ı oluşturalım. sparse=False set ettkik çünkü çıktıyı scipy.sparse metriği olarak almak istemiyorum.
dv = DictVectorizer(sparse=False)
# dv'yi object_degerler e uygulayalım
df_encoded = dv.fit_transform(df_dict)
# encoded edilmiş df'mizin satır ve sütun sayılarını döndürelim
df_encoded = pd.DataFrame(df_encoded)
```

```python
print("Shape before the encoding: ", obj_df.shape)
print("Shape after the encoding: ", df_encoded.shape)
```

* 482 new columns were added, it is too much comparing the original dataset. The more features increase, there is more chance to overfit our model. Thus, we need to apply Feature Extraction to our data and move on with the most useful features.
*  I'll plot all features and see their importance. Then, I'll measure it's prediction success. I'll name this step's outcome as df.
* I'll preprocess data with sklearn.preprocessing's MinMaxScaler as I'll use it for classification purposes.

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df_encoded)
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.gcf().set_size_inches(300, 150)
plt.xticks(features)
plt.ylabel("Variance")
plt.xlabel("PCA Feature")
plt.show()
```

```python
pca = PCA(n_components = 8)
obj_df = pca.fit_transform(df_encoded)
obj_df.shape
```

## Normalization(Min Max Scaling)

* Normalization typically means rescales the values into a range of [0,1]
* It performs better in Regression tasks.
* In Normalization, we are going to use `sklearn`s `MinMaxScaler`

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
a = scaler.fit_transform(df_encoded)
df_prep = pd.DataFrame(a)
df_prep.head()
```

## Standardization
* Standardization typically means rescales data to have a mean of 0 and a standard deviation of 1 (unit variance).
* It performs better in classification [tasks.](https://towardsdatascience.com/normalization-vs-standardization-quantitative-analysis-a91e8a79cebf#:~:text=Normalization%20typically%20means%20rescales%20the,of%201%20(unit%20variance).)

![ssc](images/ssc.png)
* In Standardization, we are going to use `Sklearn`s `StandardScaler`.
